## Bases

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, color="purple", marker="o", linestyle="--")
plt.title("Mi primer grafico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
```

## Tipos de Graficos

### DispersiÃ³n (Scatter)

```python
horas = [1, 2, 3, 4, 5]
calificaciones = [2, 3, 5, 7, 11]
plt.scatter(horas, calificaciones, s=300, c="green", marker="+")
plt.title("Calificaciones por horas de estudio")
```

### Barras

```python
panes = ["concha", "cuernito", "dona"]
ventas = [56, 90, 45]

plt.bar(panes, ventas)
```

### Pastel

```python
etiquetas = ["concha", "cuernito", "dona", "brownie"]
valores = [56, 90, 45, 30]
explotar = (0, 0, 0 , 0.3) ## separar las rebanadas
colores = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

plt.pie(valores, labels=etiquetas, explode=explotar, colors=colores) 
plt.title("Ventas de panes")

```
