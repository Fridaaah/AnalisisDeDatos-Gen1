### Promedio de ingresos brutos de cada tipo de taxi y de todos los taxis

```python
promedio_green = green["total_amount"].mean()
promedio_yellow = yellow["total_amount"].mean()

print(f"El promedio de ingreso bruto por viaje en taxis verdes es: {promedio_green:.2f} USD")
print(f"El promedio de ingreso bruto por viaje en taxis amarillos es: {promedio_yellow:.2f} USD")


promedio_ingreso_bruto = (promedio_green + promedio_yellow) / 2
print(f"El promedio de ingreso bruto de todos los taxis: {promedio_ingreso_bruto:.2f} USD")
 
```

### Qué tipo de taxis recorrieron la mayor cantidad de distancia en Enero

```python
distancia_green = green["trip_distance"].sum()
distancia_yellow = yellow["trip_distance"].sum()
distancia_fhvhv = fhvhv["trip_miles"].sum()
print(f"Distancia total recorrida por taxis verdes: {distancia_green:.2f} millas")
print(f"Distancia total recorrida por taxis amarillos: {distancia_yellow:.2f} millas")
print(f"Distancia total recorrida por FHV: {distancia_fhvhv:.2f} millas")

distancias = {
    "green": distancia_green,
    "yellow": distancia_yellow,
    "fhvhv": distancia_fhvhv
}
ganador = max(distancias, key=distancias.get)
print(f"El tipo de taxi que recorrió más distancia en enero de 2025 es: {ganador} con {distancias[ganador]:.2f} millas")
```

### Qué tipo de taxi recibe más propinas en promedio

```python
promedio_tip_green = green["tip_amount"].mean()
print(f"El promedio de propina en taxis verdes es: {promedio_tip_green:.2f} USD")

promedio_tip_yellow = yellow["tip_amount"].mean()
print(f"El promedio de propina en taxis amarillos es: {promedio_tip_yellow:.2f} USD")

promedio_tip_fhvhv = fhvhv["tips"].mean()
print(f"El promedio de propina en FHV es: {promedio_tip_fhvhv:.2f} USD")

promedio_propinas = {
    "green": promedio_tip_green,
    "yellow": promedio_tip_yellow,
    "fhvhv": promedio_tip_fhvhv
}

mas_propinas = max(promedio_propinas, key=promedio_propinas.get)
print(f"El tipo de taxi con el promedio de propinas más alto es: {mas_propinas} con {promedio_propinas[mas_propinas]:.2f} USD")
```

### Un mapa de calor que describa el comportamiento de a qué hora se tomaron todos los taxis durante los días de Enero

```python
## Renombramos columnas y agregamos columna taxy_type 
green = green.rename(columns={"lpep_pickup_datetime": "pickup_datetime"})
green["taxi_type"] = "Green"

yellow = yellow.rename(columns={"tpep_pickup_datetime": "pickup_datetime"})
yellow["taxi_type"] = "Yellow"

fhv["taxi_type"] = "FHV"

fhvhv["taxi_type"] = "HVFHV"


## Concatenamos todos los DataFrames
all_taxis = pd.concat([green, yellow, fhv, fhvhv], ignore_index=True)

## Extraemos hour y day
all_taxis["pickup_datetime"] = pd.to_datetime(all_taxis["pickup_datetime"])
all_taxis["hour"] = all_taxis["pickup_datetime"].dt.hour
all_taxis["day"] = all_taxis["pickup_datetime"].dt.day

```

```python
## Creamos el mapa de calor para todos los taxis
heatmap_all = all_taxis.groupby(["day", "hour"]).size().unstack(fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_all, cmap="YlGnBu")
plt.title("Mapa de Calor: A qué hora se toman los Taxis durante Enero")
plt.xlabel("Hora del día")
plt.ylabel("Día del mes")
plt.show()
```

### Un mapa de calor para cada tipo de taxi que reporte los días de Enero en los que más se uso su servicio

```python
## Lista de tipos de taxi únicos
taxi_types = all_taxis["taxi_type"].unique()

## Creamos un mapa de calor por cada tipo de taxi
for taxi in taxi_types:
    ## Filtrar los datos de ese taxi
    df_taxi = all_taxis[all_taxis["taxi_type"] == taxi]
    
    ## Agrupamos por día y hora
    taxi_grouped = df_taxi.groupby(["day", "hour"]).size().unstack(fill_value=0)
    
    ## Graficamos
    plt.figure(figsize=(12,6))
    sns.heatmap(taxi_grouped, cmap="YlGnBu")
    plt.title(f"{taxi} Taxi: Viajes por día y hora")
    plt.xlabel("Hora del día")
    plt.ylabel("Día de enero")
    plt.show()
```
