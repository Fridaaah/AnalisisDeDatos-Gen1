## Lectura de archivos CSV

## Selección de Datos

```python
companies = customers[customers['Index'] > 50]
gente_cuyo_nombre_tiene_a = customers[customers['First Name'].str.contains('a')]
suscritos_en_2020 = customers[customers['Subscription Date'].str.contains('2020')]
suscritos_en_2020

# iloc: Acceso por posición
primer_cliente = customers.iloc[0]
print(primer_cliente)   
```

## Eliminar Columnas

```python
customers = customers.drop(99, axis=0)
customers
```

```python
customers_sin_2telefono = customers.drop("Phone 2", axis=1)
customers_sin_2telefono
# Acceso a una columna específica
```

## Agregar columnas

```python
prebes = pd.DataFrame({
    "Nombre": ["Pepito", "Leo", "Deivi", "Bere", "Javi"],
    "Edad": [40, 21, 26, 26, 5],
    "Nota": [10, 9, 8, 7, 6]
})
prebes.loc[5] = ["Frida", 21, 10]
prebes["Sexo"] = ["M", "M", "M", "F", "M", "F"]
prebes
```

## Operaciones con datos

```python
prebes["Nota"].mean()

float(prebes["Nota"].mean())

## F-Strings
media_nota = float(prebes["Nota"].mean())

print(f"La media de las notas es: {media_nota}")
```
